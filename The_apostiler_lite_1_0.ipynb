```python
#Importando módulos
from rich.text import Text
from rich.console import Console

import pathlib
import textwrap

import typer
import time

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown

from docx import Document
from google.colab import files
```

```python
#Função para o texto de resposta
console = Console()

def print_rich(texto: str):
  """Imprime o texto usando a biblioteca rich, com quebra de linha automática."""
  texto_rich = Text(texto)
  console.print(texto_rich)

def print_chatbot(texto: str, velocidade: float = 0.02):
  """simula a impressão de texto como um chatbot """
  for letra in texto:
    typer.echo(letra, nl=False)
    time.sleep(velocidade)
  typer.echo("")
```

#Assuntos

```python
#Informa ao usuário a etapa
print_chatbot("""Vamos elaborar a lista de assuntos que serão abordados no material!
Trabalhando na lista de assuntos...\n""")

#Função para operar o loop de interação com o usuário
def obter_resposta_assuntos():
    """Obtém a resposta do usuário (s/n) e valida a entrada."""
    while True:
        resposta = input("Gostou da resposta? (s/n): ").strip().lower()
        if resposta in ("s", "n"):
            return resposta
        print("Resposta inválida. Por favor, digite 's' para sim ou 'n' para não.")

prompt_assuntos = f"""Conforme as intruções {instrucoes_sistema},
elabore uma lista simples com 8 assuntos, com breve descrição,
para uma apostila especificamente sobre {tema} para {publico_alvo} com o objetivo de {objetivo_aprendizagem}"""

#Loop de validação da resposta do modelo
while True:
    response = model.generate_content(prompt_assuntos)
    response_assuntos = response.text
    print_rich(response_assuntos) #Apresenta a resposta do modelo

    resposta_assuntos = obter_resposta_assuntos()

    if resposta_assuntos == "s":
        print("Ótimo! Vamos continuar...") #Aprovado
        break
    else:
        print("Ok, vamos gerar novamente.")
```

```python
# Separa os assuntos e seus conteúdos em variáveis
assuntos = response_assuntos.split("\n\n")
for i in range(1, 9):
    assunto = f"assunto_{i}"
    globals()[assunto] = ""
    for linha in assuntos:
        if linha.startswith(f"{i}."):
            globals()[assunto] += linha + "\n"
        elif linha.startswith(f"{i+1}."):
            break
```

```python
#Informa ao usuário a etapa
print_chatbot("""Vamos elaborar a introdução!
Trabalhando na introdução...\n""")

#Função para operar o loop de interação com o usuário
def obter_resposta_intro():
    """Obtém a resposta do usuário (s/n) e valida a entrada."""
    while True:
        resposta = input("Gostou da resposta? (s/n): ").strip().lower()
        if resposta in ("s", "n"):
            return resposta
        print("Resposta inválida. Por favor, digite 's' para sim ou 'n' para não.")

prompt_intro = f"""Conforme as intruções {instrucoes_sistema},
escreva um texto dissertativo, formal e educacional,
podendo ser formado por 3 a 5 parágrafos parágrafos,
introduzindo o tema {tema} para o público-alvo {publico_alvo}.
Destaque a importância do tema e como ele se relaciona com o objetivo de aprendizagem {objetivo_aprendizagem}"""

#Loop de validação da resposta do modelo
while True:
    response = model.generate_content(prompt_intro)
    response_intro = response.text
    print_rich(response_intro)

    resposta_intro = obter_resposta_intro()

    if resposta_intro == "s":
        print("Ótimo! Vamos continuar...") #Aprovado
        break
    else:
        print("Ok, vamos gerar novamente.")
```

#Organizando os segmentos


```python
print_chatbot("Para prosseguirmos nesta etapa, vamos resgatar a lista de assuntos. \n")

# Print de cada assunto
for i in range(1, 9):
    assunto = f"assunto_{i}"
    print_rich(globals()[assunto])

print_chatbot("Agora, vamos dividir o corpo textual do material em dois segmentos.")

#Função para operar o loop de interação com o usuário
def obter_assuntos_do_usuario(segmento):
    """Solicita ao usuário que escolha os assuntos para um segmento."""
    print(f"\nSelecione os assuntos para o {segmento}:")
    while True:
        try:
            assuntos_escolhidos = input(
                "Digite os números dos assuntos separados por vírgula (ex: 1,3,5): "
            )
            assuntos_escolhidos = [int(x.strip()) for x in assuntos_escolhidos.split(",")]
            if all(1 <= x <= 8 for x in assuntos_escolhidos):
                return assuntos_escolhidos
            else:
                print("Números inválidos. Certifique-se de inserir números de 1 a 8.")
        except ValueError:
            print("Entrada inválida. Use vírgulas para separar os números.")

while True:
    assuntos_seg_1 = obter_assuntos_do_usuario("Segmento 1") # Obtém os assuntos para o Segmento 1
    assuntos_seg_2 = obter_assuntos_do_usuario("Segmento 2") # Obtém os assuntos para o Segmento 2

    # Crie listas para armazenar o conteúdo dos segmentos
    conteudo_seg_1 = []
    conteudo_seg_2 = []

    # Adiciona o conteúdo dos assuntos escolhidos aos segmentos
    for i in assuntos_seg_1:
        conteudo_seg_1.append(globals()[f"assunto_{i}"])

    for i in assuntos_seg_2:
        conteudo_seg_2.append(globals()[f"assunto_{i}"])

    print_chatbot("\nEntão a estrutura de conteúdos distribuída entre os Segmentos ficou assim: \n")

    # Imprime o conteúdo de cada segmento (opcional - para verificar)
    print_rich("\nConteúdo do Segmento 1:")
    print_rich("\n".join(conteudo_seg_1))

    print_rich("\nConteúdo do Segmento 2:")
    print_rich("\n".join(conteudo_seg_2))

    #Loop de validação da resposta do modelo
    resposta = input("Gostou da divisão dos segmentos? (s/n): ").strip().lower()
    if resposta == "s":
        print("Ótimo! Vamos continuar...")
        break
    else:
        print("Ok, vamos refazer a divisão.\n")
```
