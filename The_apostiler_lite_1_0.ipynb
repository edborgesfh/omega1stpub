```python
#Importando m√≥dulos
from rich.text import Text
from rich.console import Console

import pathlib
import textwrap

import typer
import time

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown

from docx import Document
from google.colab import files
```

```python
#Fun√ß√£o para o texto de resposta
console = Console()

def print_rich(texto: str):
  """Imprime o texto usando a biblioteca rich, com quebra de linha autom√°tica."""
  texto_rich = Text(texto)
  console.print(texto_rich)

def print_chatbot(texto: str, velocidade: float = 0.02):
  """simula a impress√£o de texto como um chatbot """
  for letra in texto:
    typer.echo(letra, nl=False)
    time.sleep(velocidade)
  typer.echo("")
```

#Assuntos

```python
#Informa ao usu√°rio a etapa
print_chatbot("""Vamos elaborar a lista de assuntos que ser√£o abordados no material!
Trabalhando na lista de assuntos...\n""")

#Fun√ß√£o para operar o loop de intera√ß√£o com o usu√°rio
def obter_resposta_assuntos():
    """Obt√©m a resposta do usu√°rio (s/n) e valida a entrada."""
    while True:
        resposta = input("Gostou da resposta? (s/n): ").strip().lower()
        if resposta in ("s", "n"):
            return resposta
        print("Resposta inv√°lida. Por favor, digite 's' para sim ou 'n' para n√£o.")

prompt_assuntos = f"""Conforme as intru√ß√µes {instrucoes_sistema},
elabore uma lista simples com 8 assuntos, com breve descri√ß√£o,
para uma apostila especificamente sobre {tema} para {publico_alvo} com o objetivo de {objetivo_aprendizagem}"""

#Loop de valida√ß√£o da resposta do modelo
while True:
    response = model.generate_content(prompt_assuntos)
    response_assuntos = response.text
    print_rich(response_assuntos) #Apresenta a resposta do modelo

    resposta_assuntos = obter_resposta_assuntos()

    if resposta_assuntos == "s":
        print("√ìtimo! Vamos continuar...") #Aprovado
        break
    else:
        print("Ok, vamos gerar novamente.")
```

```python
# Separa os assuntos e seus conte√∫dos em vari√°veis
assuntos = response_assuntos.split("\n\n")
for i in range(1, 9):
    assunto = f"assunto_{i}"
    globals()[assunto] = ""
    for linha in assuntos:
        if linha.startswith(f"{i}."):
            globals()[assunto] += linha + "\n"
        elif linha.startswith(f"{i+1}."):
            break
```

#Introdu√ß√£o

```python
#Informa ao usu√°rio a etapa
print_chatbot("""Vamos elaborar a introdu√ß√£o!
Trabalhando na introdu√ß√£o...\n""")

#Fun√ß√£o para operar o loop de intera√ß√£o com o usu√°rio
def obter_resposta_intro():
    """Obt√©m a resposta do usu√°rio (s/n) e valida a entrada."""
    while True:
        resposta = input("Gostou da resposta? (s/n): ").strip().lower()
        if resposta in ("s", "n"):
            return resposta
        print("Resposta inv√°lida. Por favor, digite 's' para sim ou 'n' para n√£o.")

prompt_intro = f"""Conforme as intru√ß√µes {instrucoes_sistema},
escreva um texto dissertativo, formal e educacional,
podendo ser formado por 3 a 5 par√°grafos par√°grafos,
introduzindo o tema {tema} para o p√∫blico-alvo {publico_alvo}.
Destaque a import√¢ncia do tema e como ele se relaciona com o objetivo de aprendizagem {objetivo_aprendizagem}"""

#Loop de valida√ß√£o da resposta do modelo
while True:
    response = model.generate_content(prompt_intro)
    response_intro = response.text
    print_rich(response_intro)

    resposta_intro = obter_resposta_intro()

    if resposta_intro == "s":
        print("√ìtimo! Vamos continuar...") #Aprovado
        break
    else:
        print("Ok, vamos gerar novamente.")
```

#Organizando os segmentos


```python
print_chatbot("Para prosseguirmos nesta etapa, vamos resgatar a lista de assuntos. \n")

# Print de cada assunto
for i in range(1, 9):
    assunto = f"assunto_{i}"
    print_rich(globals()[assunto])

print_chatbot("Agora, vamos dividir o corpo textual do material em dois segmentos.")

#Fun√ß√£o para operar o loop de intera√ß√£o com o usu√°rio
def obter_assuntos_do_usuario(segmento):
    """Solicita ao usu√°rio que escolha os assuntos para um segmento."""
    print(f"\nSelecione os assuntos para o {segmento}:")
    while True:
        try:
            assuntos_escolhidos = input(
                "Digite os n√∫meros dos assuntos separados por v√≠rgula (ex: 1,3,5): "
            )
            assuntos_escolhidos = [int(x.strip()) for x in assuntos_escolhidos.split(",")]
            if all(1 <= x <= 8 for x in assuntos_escolhidos):
                return assuntos_escolhidos
            else:
                print("N√∫meros inv√°lidos. Certifique-se de inserir n√∫meros de 1 a 8.")
        except ValueError:
            print("Entrada inv√°lida. Use v√≠rgulas para separar os n√∫meros.")

while True:
    assuntos_seg_1 = obter_assuntos_do_usuario("Segmento 1") # Obt√©m os assuntos para o Segmento 1
    assuntos_seg_2 = obter_assuntos_do_usuario("Segmento 2") # Obt√©m os assuntos para o Segmento 2

    # Crie listas para armazenar o conte√∫do dos segmentos
    conteudo_seg_1 = []
    conteudo_seg_2 = []

    # Adiciona o conte√∫do dos assuntos escolhidos aos segmentos
    for i in assuntos_seg_1:
        conteudo_seg_1.append(globals()[f"assunto_{i}"])

    for i in assuntos_seg_2:
        conteudo_seg_2.append(globals()[f"assunto_{i}"])

    print_chatbot("\nEnt√£o a estrutura de conte√∫dos distribu√≠da entre os Segmentos ficou assim: \n")

    # Imprime o conte√∫do de cada segmento (opcional - para verificar)
    print_rich("\nConte√∫do do Segmento 1:")
    print_rich("\n".join(conteudo_seg_1))

    print_rich("\nConte√∫do do Segmento 2:")
    print_rich("\n".join(conteudo_seg_2))

    #Loop de valida√ß√£o da resposta do modelo
    resposta = input("Gostou da divis√£o dos segmentos? (s/n): ").strip().lower()
    if resposta == "s":
        print("√ìtimo! Vamos continuar...")
        break
    else:
        print("Ok, vamos refazer a divis√£o.\n")
```
